```python
def noisy(noise_typ,image):
    if noise_typ == "gauss":
        row,col,ch= image.shape
        mean = 0
        var = 15
        sigma = var**0.5
        print(sigma)
        gauss = np.random.normal(mean,sigma,(row,col,ch))
        gauss = gauss.reshape(row,col,ch)
        noisy = image + gauss + gauss + gauss
        return noisy
    elif noise_typ == "s&p":
        row,col,ch = image.shape
        s_vs_p = 0.5
        amount = 0.004
        out = np.copy(image)
        # Salt mode
        num_salt = np.ceil(amount * image.size * s_vs_p)
        coords = [np.random.randint(0, i - 1, int(num_salt))
                for i in image.shape]
        out[coords] = 1

        # Pepper mode
        num_pepper = np.ceil(amount* image.size * (1. - s_vs_p))
        coords = [np.random.randint(0, i - 1, int(num_pepper))
                for i in image.shape]
        out[coords] = 0
        return out
    elif noise_typ == "poisson":
        vals = len(np.unique(image))
        vals = 2 ** np.ceil(np.log2(vals))
        noisy = np.random.poisson(image * vals) / float(vals)
        return noisy
    elif noise_typ =="speckle":
        row,col,ch = image.shape
        gauss = np.random.randn(row,col,ch)
        gauss = gauss.reshape(row,col,ch)
        noisy = image + image * gauss
        return noisy
```

```python

def getTrainingData():

    transform = transforms.Compose([transforms.ToTensor(),])#transforms images to tensors

    img_dir = "output_top"#images
    gt_dir = "output_gts_for_participants"#labels

    data_path1 = os.path.join(img_dir, '**', '*.jpg')
    data_path2 = os.path.join(gt_dir, '**', '*.jpg')#labels path
    files1 = sorted(glob.glob(data_path1), key=os.path.getmtime)#finds files that have the same datapath1 and sorts based on last time modified
    files2 = sorted(glob.glob(data_path2), key=os.path.getmtime)
    data_input = []
    data_label = []

    for f1 in files1:
        img = cv2.imread(f1)#read image
        img = transform(img)#turn it into tensor
        img = img.unsqueeze(0)#add dimension
        data_input.append(img)#add it to list

    for f2 in files2:
        img = cv2.imread(f2)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        img = torch.from_numpy(img.astype(int))
        img = img.long()
        img = img.unsqueeze(0)
        data_label.append(img)

    tuple_list = []
    for x in range(len(data_input)):
        pair = (data_input[x], data_label[x])
        tuple_list.append(pair)

    return tuple_list
```

```python
def getTestData(img_size=320):
    transform = transforms.Compose([transforms.ToTensor(),])
    img_dir = "Edits"
    data_path1 = os.path.join(img_dir,'*.jpg')#takes path of png ?????
    files1 = sorted(glob.glob(data_path1), key=os.path.getmtime)#finds all images with the pattern from datapath1 sorted
    data_input = []
    for f1 in files1:
        img = cv2.imread(f1)
        img = img[0:img_size,0:img_size,:] # crops image ??
        print(img.shape)
        cv2.imwrite("result\\original.png", img)
        #scipy.misc.imsave('test\pic\gradient.png', arr_gtruth)
        img = noisy("gauss", img)# Adding noise to the image
        cv2.imwrite("result\\noisy.png", img)
        img = transform(img)#after all chnages turn it into a tensor
        img = img.float()
        img = img.unsqueeze(0)# add dimension
        data_input.append(img)#add to list

    return data_input
```

```python
def getBatchData(tuple_list, batch_size = 1):
    batch_tuple_list = []
    shuffle(tuple_list)
    x = 0
    while x < len(tuple_list):
        arg1 = tuple_list[x][0]
        arg2 = tuple_list[x][1]
        x += 1
        for i in range(batch_size-1):
            if x >= len(tuple_list):
                break

            arg3 = tuple_list[x][0]
            arg1 = torch.cat((arg1, arg3),0)

            arg4 = tuple_list[x][1]
            arg2 = torch.cat((arg2, arg4),0)

            x +=1

        pair = (arg1, arg2)
        batch_tuple_list.append(pair)

    return batch_tuple_list
```

```python
def displayGenerated(gen, i, k, mode="standard"):
    bat = gen.shape[0]
    l = gen.shape[2]
    w = gen.shape[3]
    gen = gen.cpu().numpy()

    for z in range(bat):
        dis = np.zeros((l, w, 3), dtype=np.uint8)
        for x in range(l):
            for y in range(w):
                if gen.item((z, 0, x, y)) == 0:
                    dis[x,y,0] = 255
                    if mode == "difference":
                        dis[x,y,1] = 255
                        dis[x,y,2] = 255
                if gen.item((z, 0, x, y)) == 1:
                    dis[x,y,1] = 255
                if gen.item((z, 0, x, y)) == 2:
                    dis[x,y,2] = 255
       # scipy.misc.imsave('result\\argmax\\argmax_{}_{}_{}.png'.format(i,k,z), dis)
        imageio.imwrite('./result/truth_{}_{}_{}.png'.format(i,k,z), dis)
```

```python
def displayTruth(truth, i, k):
    bat = truth.shape[0]
    l = truth.shape[1]
    w = truth.shape[2]

    truth = truth.cpu().numpy()
    for z in range(bat):
        dis = np.zeros((l, w), dtype=np.uint8)
        for x in range(l):
            for y in range(w):
                if truth.item((z,x,y)) == 0:
                    dis[x,y] = 0
                if truth.item((z,x,y)) == 1:
                    dis[x,y] = 100
                if truth.item((z,x,y)) == 2:
                    dis[x,y] = 200
        #scipy.misc.imsave('result\\truth_{}_{}_{}.png'.format(i,k,z), dis)
        imageio.imwrite('./result/truth_{}_{}_{}.png'.format(i,k,z), dis)
```
